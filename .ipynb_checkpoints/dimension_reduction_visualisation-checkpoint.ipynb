```python
import pandas as pd

path= "/content/drive/MyDrive/phishing_urls/embeddings.pkl"
df = pd.read_pickle(path)
df
```

```python
import numpy as np

# Convert string representations to actual lists
df['Embeddings'] = df['Embeddings'].apply(eval)

# Convert lists to numpy arrays
df['Embeddings'] = df['Embeddings'].apply(np.array)

# Check the type of the first element
print(type(df['Embeddings'][0]))
```

## Run UMAP on subset

```python
import pandas as pd

# Filter 'good' and 'bad' samples and create a smaller subset
good_samples = df[df['Label'] == 'good'].iloc[:2500]
bad_samples = df[df['Label'] == 'bad'].iloc[:2500]
subset_df = pd.concat([good_samples, bad_samples], ignore_index=True)
```

Try with low number of neighbours to prioritize local structure

```python
import umap
import matplotlib.pyplot as plt

# Extract the embeddings from the subset DataFrame
embeddings = subset_df['Embeddings'].tolist()

# Initialize UMAP model with desired parameters
umap_model = umap.UMAP(n_neighbors=10, n_components=2, random_state=42)

# Fit UMAP to the embeddings
umap_embeddings = umap_model.fit_transform(embeddings)

# Create a DataFrame for visualization
umap_df = pd.DataFrame(umap_embeddings, columns=['UMAP1', 'UMAP2'])

# Add labels to the subset DataFrame
umap_df['Label'] = subset_df['Label']

# Define colors for 'bad' and 'good' labels
label_colors = {'bad': 'red', 'good': 'green'}

# Visualize the UMAP embeddings with colors
plt.figure(figsize=(10, 6))
for label, color in label_colors.items():
    label_df = umap_df[umap_df['Label'] == label]
    plt.scatter(label_df['UMAP1'], label_df['UMAP2'], s=5, c=color, label=label)

plt.title('UMAP Visualization of Embeddings')
plt.xlabel('UMAP1')
plt.ylabel('UMAP2')
plt.legend()
plt.show()
```
